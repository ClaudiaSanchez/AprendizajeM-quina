```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas
```

```python
df = pandas.read_csv('adult.data.csv')
#df.columns # columnas
#df.values  # datos
print(df.values.shape)
print(df)
```

```python
feature = df.values[:,-2]
feature_values, feature_counts = np.unique(feature,return_counts=True)
for i in range(len(feature_values)):
    print(feature_values[i],feature_counts[i])
```

```python
plt.figure()
plt.bar(np.arange(len(feature_values)), feature_counts)
plt.xticks(np.arange(len(feature_values)),feature_values,rotation='vertical')
plt.show()
```

```python
plt.figure()
plt.pie(feature_counts,labels=feature_values)
plt.show()
```

```python
print(df.columns)
print(df.columns[3])
print(df.columns[9])
```

```python
genero = df.values[:,9]
estudios = df.values[:,3]
print(np.unique(genero,return_counts=True))
```

```python
f_et_estudios,f_fr_estudios = np.unique(estudios[genero==' Female'],return_counts=True)
print(f_et_estudios,f_fr_estudios)
m_et_estudios,m_fr_estudios = np.unique(estudios[genero==' Male'],return_counts=True)
print(m_et_estudios,m_fr_estudios)

plt.figure()
plt.subplot(1,2,1)
plt.title(' Female')
plt.pie(f_fr_estudios,labels=f_et_estudios)
plt.subplot(1,2,2)
plt.title(' Male')
plt.pie(m_fr_estudios,labels=m_et_estudios)
plt.show()
```
