```python
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
X = iris.data[:, :2]  
y = iris.target

plt.figure()
plt.plot(X[:,0],X[:,1],'.')
plt.show()
```

```python
model = GaussianMixture(n_components=2)
model.fit(X)
g = model.predict(X)
print(len(g))
print(g)

plt.figure()
plt.plot(X[g==0,0],X[g==0,1],'.')
plt.plot(X[g==1,0],X[g==1,1],'.')
plt.show()
```

```python
model = KMeans(n_clusters=2)
model.fit(X)
g = model.predict(X)
print(len(g))
print(g)

plt.figure()
plt.plot(X[g==0,0],X[g==0,1],'.')
plt.plot(X[g==1,0],X[g==1,1],'.')
plt.show()
```
