```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Opci√≥n 1 para leer y mostrar imagen
image = cv2.imread('mariposa.jpg')
cv2.imshow('UP opencv',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Opci√≥n 2 para leer y mostrar imagen
image = cv2.imread('mariposa.jpg') # La imagen se lee BGR
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure()
plt.imshow( image )
plt.show()
```

```python
# CAMBIOS DE ESPACIOS DE COLOR
image = cv2.imread('mariposa.jpg') # La imagen se lee BGR
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
image_hsv = cv2.cvtColor(image,cv2.COLOR_RGB2HSV)
plt.figure()
plt.subplot(1,3,1)
plt.imshow( image )
plt.subplot(1,3,2)
plt.imshow(image_gray,cmap='gray')
plt.subplot(1,3,3)
plt.imshow(image_hsv)
plt.show()
```

```python
# MASCARA PARA DETECTAR AZULES
image = cv2.imread('mariposa.jpg') # La imagen se lee BGR
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

lower = np.array([90,50,50])
upper = np.array([130,255,255])
mask = cv2.inRange(image_hsv,lower,upper)
print('mask',mask.shape)
plt.imshow(mask,cmap='gray')
plt.show()

image_mask = cv2.bitwise_and(image,image,mask=mask)
print('image_mask',image_mask.shape)
plt.imshow( cv2.cvtColor(image_mask,cv2.COLOR_BGR2RGB ))
plt.show()
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
image_gray = np.copy(image)
image_gray[:,:,0] = gray
image_gray[:,:,1] = gray
image_gray[:,:,2] = gray
print('image-gray',image_gray.shape)
plt.imshow(gray,cmap='gray')
plt.show()

gray_mask = cv2.bitwise_and(image_gray,image_gray,mask=~mask)
print('gray_mask',gray_mask.shape)
plt.imshow(gray_mask)
plt.show()

image_res = image_mask + gray_mask
print('image_res',image_res.shape)
plt.imshow( cv2.cvtColor(image_res,cv2.COLOR_BGR2RGB ))
plt.show()
```
