# PRINCIPAL COMPONENT ANALYSIS

```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
```

```python
X,y = load_digits(return_X_y = True)
print('X',X.shape)
print('y',y.shape)
```

```python
idx = 750
sample = X[idx,:]
print('Digito:', y[idx])
print('imagen vectorizada:',sample)
img_sample = np.reshape(sample,(8,8))
print('imagen valores:',img_sample)
plt.imshow(img_sample,cmap='gray')
plt.show()
```

### Dimensionality reduction

```python
model = PCA()
model.fit(X)
XX = model.transform(X)
```

```python
print('Datos originales:','X',X.shape)
print('Datos proyectados:','XX',XX.shape)
```

```python
idx = np.random.permutation(len(X))
idx = idx[:100]
```

```python
plt.figure()
plt.subplot(1,2,1)
plt.plot(X[idx,0],X[idx,1],'.')
for i in idx:
    plt.text(X[i,0],X[i,1],y[i])

plt.subplot(1,2,2)
plt.plot(XX[idx,0],XX[idx,1],'.')
for i in idx:
    plt.text(XX[i,0],XX[i,1],y[i])
plt.show()
```

## Explained variance

```python
print(np.round(model.explained_variance_ratio_,2))
print(sum(model.explained_variance_ratio_))
acc_exp_var = np.cumsum(model.explained_variance_ratio_)
print('acc_exp_var',np.round(acc_exp_var,2))
```

```python
plt.plot(acc_exp_var)
plt.show()
for i in range(len(acc_exp_var )):
    print(i,acc_exp_var[i])
    
```
