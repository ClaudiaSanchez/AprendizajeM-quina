```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
X,y = load_iris(return_X_y=True)
print('X',X.shape)
print('y',np.unique(y))
```

```python
X = X[:,[0,2]]
print('X',X.shape)
```

```python
plt.figure()
plt.plot(X[y==0,0],X[y==0,1],'.',label='class 0')
plt.plot(X[y==1,0],X[y==1,1],'.',label='class 1')
plt.plot(X[y==2,0],X[y==2,1],'.',label='class 2')
plt.legend()
plt.show()
```

```python
modelPCA = PCA()
modelPCA.fit(X)
XPCA = modelPCA.transform(X)

modelLDA = LinearDiscriminantAnalysis()
modelLDA.fit(X,y)
XLDA = modelLDA.transform(X)

print('X',X.shape)
print('XPCA',XPCA.shape)
print('XLDA',XLDA.shape)
```

```python
plt.figure()
plt.subplot(2,2,1)
plt.title('X')
plt.plot(X[y==0,0],X[y==0,1],'.',label='class 0')
plt.plot(X[y==1,0],X[y==1,1],'.',label='class 1')
plt.plot(X[y==2,0],X[y==2,1],'.',label='class 2')

plt.subplot(2,2,3)
plt.title('PCA')
plt.plot(XPCA[y==0,0],XPCA[y==0,1],'.',label='class 0')
plt.plot(XPCA[y==1,0],XPCA[y==1,1],'.',label='class 1')
plt.plot(XPCA[y==2,0],XPCA[y==2,1],'.',label='class 2')

plt.subplot(2,2,4)
plt.title('LDA')
plt.plot(XLDA [y==0,0],XLDA[y==0,1],'.',label='class 0')
plt.plot(XLDA [y==1,0],XLDA[y==1,1],'.',label='class 1')
plt.plot(XLDA [y==2,0],XLDA[y==2,1],'.',label='class 2')

plt.show()
```
