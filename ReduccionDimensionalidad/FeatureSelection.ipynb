```python
import numpy as np
import matplotlib.pyplot as plt
import pandas
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error 
from sklearn.model_selection import train_test_split
```

```python
df=pandas.read_csv('winequality-red.csv',sep=';')
df
```

```python
# Matriz de correlación
CORR = np.corrcoef(df.values.transpose())
dfCORR = pandas.DataFrame(CORR,index=df.columns,columns=df.columns)
print(dfCORR)
```

```python
corr = np.abs(CORR[:,-1])
for i in range(len(corr)):
    print(df.columns[i],corr[i])
idx = np.argsort(corr)
idx = list(idx)
idx.reverse()
idx = np.array(idx)
names_features = df.columns[idx]
corr = corr[idx]
for i in range(len(corr)):
    print(idx[i],names_features[i],corr[i])
```

```python
# Entrenamiento y prueba
X = df.values[:,:-1]
Y = df.values[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.3)
```

## Modelo con 1 variable

```python
XXtrain = np.reshape(X_train[:,10], (len(X_train),1))
XXtest = np.reshape(X_test[:,10],(len(X_test),1))
nfeatures = df.columns[10]
print(nfeatures,'XXtrain',XXtrain.shape)

model = SVR()
model.fit(XXtrain,y_train)
ypred = model.predict(XXtest)
print('MAE',mean_absolute_error(y_test,ypred))
```

## Modelo con 2 variables

```python
XXtrain = X_train[:,[10,1]]
XXtest = X_test[:,[10,1]]
nfeatures = df.columns[[10,1]]
print(nfeatures,'XXtrain',XXtrain.shape)

model = SVR()
model.fit(XXtrain,y_train)
ypred = model.predict(XXtest)
print('MAE',mean_absolute_error(y_test,ypred))
```

## Modelo con todas las variables

```python
model = SVR()
model.fit(X_train,y_train)
ypred = model.predict(X_test)
print('MAE',mean_absolute_error(y_test,ypred))
```
