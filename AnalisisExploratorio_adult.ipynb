```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas
```

```python
df = pandas.read_csv('adult.data.txt')
#df.columns # columnas
#df.values  # datos
print(df.values.shape)
print(df)
```

```python
feature = df.values[:,-2]
feature_values, feature_counts = np.unique(feature,return_counts=True)
for i in range(len(feature_values)):
    print(feature_values[i],feature_counts[i])
```

```python
plt.figure()
plt.bar(np.arange(len(feature_values)), feature_counts)
plt.xticks(np.arange(len(feature_values)),feature_values,rotation='vertical')
plt.show()
```

```python
plt.figure()
plt.pie(feature_counts,labels=feature_values)
plt.show()
```
