```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas
```

```python
df = pandas.read_csv('winequality-red.csv',sep=';')
print(df)
```

```python
feature = df.values[:,0]
name = df.columns[0]
plt.figure()
plt.title(name)
plt.hist(feature)
plt.show()
```

```python
feature = df.values[:,0]
name = df.columns[0]
plt.figure()
plt.title(name)
plt.boxplot(feature)
plt.show()
```

```python
feature = df.values[:,0]
name = df.columns[0]
plt.figure()
plt.subplot(1,2,1)
plt.boxplot(feature)
plt.subplot(1,2,2)
plt.hist(feature)
plt.suptitle(name)
plt.show()
```
